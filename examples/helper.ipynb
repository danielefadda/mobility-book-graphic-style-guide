## 1. Palette di Colori

````python
# Ottieni una palette base in ordine (chiaro → scuro)
teal_palette = theme.get_palette('teal')
print(f"Teal palette ({len(teal_palette)} colori):")
print(f"  Più chiaro: {teal_palette[0]}")
print(f"  Più scuro: {teal_palette[-1]}")
print(f"  Completa: {teal_palette}")

# Palette disponibili
print(f"\nPalette disponibili: {', '.join(theme.list_palettes())}")
````

## 2. Colori Semantici

````python
# Accesso rapido ai colori principali
primary = theme.get_color('primary')
accent = theme.get_color('accent')

print(f"Brand colors:")
print(f"  Primary: {primary}")
print(f"  Accent: {accent}")

# Colori intent (esercizio, avvertimento, etc)
exercise = theme.get_color('exercise', section='intent')
warning = theme.get_color('warning', section='intent')

print(f"\nIntent colors:")
print(f"  Exercise: {exercise}")
print(f"  Warning: {warning}")

# Colori testo
text_primary = theme.get_color('text_primary')
link = theme.get_color('link')

print(f"\nText colors:")
print(f"  Primary: {text_primary}")
print(f"  Link: {link}")
````

## 3. Configurazioni Font

````python
# Font props pronti per matplotlib
title_font = theme.get_font_props(family='sans', size='xl', weight='bold')
print(f"Title font (sans, xl, bold):")
for key, val in title_font.items():
    print(f"  {key}: {val}")

# Altre combinazioni
body_font = theme.get_font_props(size='m', weight='regular')
label_font = theme.get_font_props(size='s', weight='regular')

print(f"\nBody font: {body_font}")
print(f"Label font: {label_font}")

# Scopri le opzioni disponibili
print(f"\nFont families: {', '.join(theme.list_fonts())}")
print(f"Font sizes: {', '.join(theme.list_font_sizes())}")
print(f"Font weights: {', '.join(theme.list_font_weights())}")
````

## 4. Sequenze per Grafici

````python
# Sequenze di colori per grafici Altair/Matplotlib
rust_seq = theme.get_chart_sequence('rust')
print(f"Rust sequence ({len(rust_seq)} colori):")
print(f"  {rust_seq}")

# Tutte le sequenze disponibili
schemes = theme.list_chart_schemes()
print(f"\nSequenziali disponibili:")
print(f"  {', '.join(schemes.get('sequential', []))}")
print(f"\nDivergenti disponibili:")
print(f"  {', '.join(schemes.get('divergent', []))}")
````

## Casi d'uso Pratici: Matplotlib

````python
import matplotlib.pyplot as plt
import numpy as np

# Setup tema matplotlib
mbs.apply_matplotlib_theme()

# Crea un grafico usando i colori dal theme
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: usa il colore semantico "exercise"
x = np.linspace(0, 10, 100)
ax1.plot(x, np.sin(x), color=theme.get_color('exercise'), linewidth=2)
ax1.set_title('Usando get_color("exercise")', fontdict=theme.get_font_props(size='l', weight='bold'))
ax1.grid(True, alpha=0.3)

# Plot 2: usa una palette per una heatmap
from matplotlib.colors import ListedColormap
palette = theme.get_palette('teal')
cmap = ListedColormap(palette)
data = np.random.rand(10, 10)
im = ax2.imshow(data, cmap=cmap, aspect='equal')
ax2.set_title('Usando get_palette("teal")', fontdict=theme.get_font_props(size='l', weight='bold'))
plt.colorbar(im, ax=ax2)

plt.tight_layout()
plt.show()
````

## Casi d'uso Pratici: Altair

````python
import altair as alt
import pandas as pd

# Setup tema altair
mbs.enable_altair_theme()

# Crea un grafico con sequenza teal
data = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': [10, 23, 15, 30, 22]
})

teal_colors = theme.get_chart_sequence('teal')

chart = alt.Chart(data).mark_bar().encode(
    x='Category:N',
    y='Value:Q',
    color=alt.Color('Category:N', scale=alt.Scale(range=teal_colors[:5]))
).properties(
    title='Grafico con theme.get_chart_sequence("teal")',
    width=500,
    height=300
)

chart
````

## Genera HTML di Visualizzazione (Esterno)

È disponibile uno script Python per generare un file HTML autonomo con la visualizzazione interattiva dei colori.
Questo file si aggiorna automaticamente quando cambiano i design token!

````python
from mobility_book_style.theme_visualizer import save_theme_html

# Genera l'HTML di visualizzazione
html_path = save_theme_html()
print(f"✅ File HTML generato: {html_path}")
print(f"📁 Percorso: {html_path.as_posix()}")
print(f"\n💡 Il file si aggiorna automaticamente quando cambiano i design token!")
print(f"   Basta eseguire:")
print(f"   python scripts/generate_theme_visualization.py")
````

### Come Funziona

Lo script Python `theme_visualizer.py` genera HTML **dinamico** usando le funzioni del modulo `theme`:

1. **Legge i token reali** - Non codifica i colori manualmente
2. **Usa `theme.get_palette()`** per ottenere tutte le palette
3. **Usa `theme.get_color()`** per i colori semantici
4. **Usa `theme.get_chart_sequence()`** per le sequenze

**Risultato:** Se cambiano i design token → basta rigenerare l'HTML e vedrà automaticamente le nuove palette!
